```python
import os
import pandas as pd
```

```python
# Define the directory where the files are located
base_dir = './'  # Replace with the path to your folder

# Function to recursively find and convert all .xlsx files to .csv
def convert_xlsx_to_csv(base_dir):
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file.endswith('.xlsx'):
                xlsx_path = os.path.join(root, file)
                csv_path = xlsx_path.replace('.xlsx', '.csv')
                try:
                    # Read the excel file and convert it to csv
                    df = pd.read_excel(xlsx_path)
                    df.to_csv(csv_path, index=False)
                    print(f"Converted {xlsx_path} to {csv_path}")
                except Exception as e:
                    print(f"Error converting {xlsx_path}: {e}")
```

```python
# Call the function to convert all xlsx to csv
convert_xlsx_to_csv(base_dir)
```
